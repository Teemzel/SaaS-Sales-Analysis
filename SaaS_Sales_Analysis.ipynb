Descriptive Analysis: What Happened?
Sales and profit trends over time: Which months, quarters, or years show the highest growth?

```python
df['Year'] = df['Order Date'].dt.year
df['Quarter'] = df['Order Date'].dt.to_period('Q')
df['Month'] = df['Order Date'].dt.to_period('M')
df
```

```python
time_trend = df.groupby('Month')[['Sales', 'Profit']].sum().reset_index()
time_trend['Month'] = time_trend['Month'].dt.to_timestamp()
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 5))
plt.plot(time_trend['Month'], time_trend['Sales'], label='Gross Sales', color='blue', linewidth=2)
plt.plot(time_trend['Month'], time_trend['Profit'], label='Net Profit', color='green', linewidth=2)
plt.title('Monthly Sales and Profit Trends Over Time', fontsize=14)
plt.xlabel('Order Date')
plt.ylabel('Financial Amount ($)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
discount_diag = df.groupby('Discount')[['Sales', 'Profit']].mean().reset_index()
discount_diag['Profit_Margin_%'] = (discount_diag['Profit'] / discount_diag['Sales']) * 100
discount_diag['Sales'] = discount_diag['Sales'].map('{:.2f}'.format)
discount_diag['Profit'] = discount_diag['Profit'].map('{:.2f}'.format)
discount_diag['Profit_Margin_%'] = discount_diag['Profit_Margin_%'].map('{:.2f}'.format)
print("Average Profit Margin by Discount Level")
display(discount_diag.sort_values(by='Discount'))
```

```python
# Visualize the Discount-Profit Tipping Point
discount_diag['Profit_Margin_%'] = pd.to_numeric(discount_diag['Profit_Margin_%'], errors='coerce')
discount_diag = discount_diag.dropna(subset=['Profit_Margin_%'])
plt.figure(figsize=(10, 5))
sns.lineplot(data=discount_diag, x='Discount', y='Profit_Margin_%', marker='o', color='red', linewidth=2)
plt.axhline(0, color='black', linestyle='--', alpha=0.7)
for idx, row in discount_diag.iterrows():
    x = row['Discount']
    y = row['Profit_Margin_%']
    # Push labels slightly above positive points, and below negative points to prevent overlapping the line
    offset = 4 if y >= 0 else -12
    
    plt.annotate(
        f"{y:.1f}%", 
        (x, y),
        textcoords="offset points", 
        xytext=(0, offset), 
        ha='center', 
        fontsize=9, 
        fontweight='bold',
        color='darkred' if y < 0 else 'darkgreen'
    )
plt.title('The Tipping Point: How Discounts Destroy Profit Margin', fontsize=14)
plt.xlabel('Discount Percentage Applied')
plt.ylabel('Average Profit Margin (%)')
#plt.ylim(discount_diag['Profit_Margin_%'].min() - 20, discount_diag['Profit_Margin_%'].max() + 20)
plt.tight_layout()
plt.show()
```

```python
# 3. Identify your top profit-bleeding orders
bleeding_orders = df[df['Profit'] < 0].sort_values(by='Profit').head(10)
print("\nTop 10 Most Unprofitable Orders - Diagnostic Sample")
display(bleeding_orders[['Order ID', 'Product', 'Sales', 'Discount', 'Profit']])
```

```python
Profit_orders = df[df['Profit'] > 0].sort_values(by='Profit', ascending=False).head(10)
print("\nTop 10 Most profitable Orders - Diagnostic Sample")
display(Profit_orders[['Order ID', 'Product', 'Sales', 'Discount', 'Profit']])
```

```python
# Top 5 Subregions by Profit
top_subregions = df.groupby('Subregion')[['Sales', 'Profit']].sum().sort_values(by='Profit', ascending=False).head(5)
formatted_subregions = top_subregions.style.format({
    'Sales': '${:,.2f}',
    'Profit': '${:,.2f}'
})
formatted_subregions
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

sns.barplot(data=top_subregions.reset_index(), x='Profit', y='Subregion', hue='Subregion', palette='viridis', ax=ax)

for container in ax.containers:
    ax.bar_label(
        container, 
        labels=[f"${x:,.0f}" for x in container.datavalues],          # Formats numbers as currency (e.g., $150,000)
        padding=8,              # Pushes the text slightly right of the bar end
        fontsize=10, 
        fontweight='bold',
        color='#2D3748'
    )
plt.title('Top 5 Most Profitable Subregions', fontsize=14)
plt.xlabel('Net Profit ($)', fontsize=11, labelpad=10, fontweight='bold', color='#2D3748')
plt.ylabel('Sub Regions', fontsize=11, labelpad=10, fontweight='bold', color='#2D3748')

ax.set_xlim(right=top_subregions['Profit'].max() * 1.15)

plt.tight_layout()
plt.savefig('2_top_profitable_subregions.png')
plt.show()
```

```python
# Year-over-Year Growth summary table
yoy_metrics = df.groupby('Year')[['Sales', 'Profit']].sum()
print("\nYoY Financial Performance Summary")
display(yoy_metrics)
```

```python
print(" STRATEGIC FINANCIAL METRICS ")
# Calculate Net Profit Margin Percentage
yoy_metrics['Profit_Margin_%'] = (yoy_metrics['Profit'] / yoy_metrics['Sales']) * 100

# Calculate Year-over-Year Percentage Growth
yoy_metrics['Sales_YoY_Growth_%'] = yoy_metrics['Sales'].pct_change() * 100
yoy_metrics['Profit_YoY_Growth_%'] = yoy_metrics['Profit'].pct_change() * 100

# Format the dataframe
formatted_yoy = yoy_metrics.style.format({
    'Sales': '${:,.2f}',
    'Profit': '${:,.2f}',
    'Profit_Margin_%': '{:.2f}%',
    'Sales_YoY_Growth_%': '{:+.2f}%',
    'Profit_YoY_Growth_%': '{:+.2f}%'
}, na_rep='-')

display(formatted_yoy)
```

```python
# Top-Performing Geographic Zones
# Best regions by Revenue, Profit and Profit Margin 
geo_region = df.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
geo_region['Profit_Margin_%'] = (geo_region['Profit'] / geo_region['Sales']) * 100
print("Top Regions by Sales")
formated_geo_region = geo_region.style.format({
    'Sales': '${:,.2f}',
    'Profit': '${:,.2f}',
    'Profit_Margin_%': '{:.2f}%'
})
display(formated_geo_region)
```

```python
# Industry Segment Spending
segment_spending = df.groupby(['Segment', 'Industry'])['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
segment_spending['Sales'] = segment_spending['Sales'].map('${:,.2f}'.format)
display(segment_spending[['Segment', 'Industry', 'Sales']])
```

```python
# Pivot and sort data by TOTAL sales
pivot_df = df.groupby(['Industry', 'Segment'])['Sales'].sum().unstack()
pivot_df['Total'] = pivot_df.sum(axis=1)
pivot_df = pivot_df.sort_values(by='Total', ascending=True)
totals = pivot_df['Total'].values  # Save totals for the end labels
pivot_df = pivot_df.drop(columns='Total')

fig, ax = plt.subplots(figsize=(12, 7))
colors = ['#2B6CB0', '#DD6B20', '#48BB78']
pivot_df.plot(kind='barh', stacked=True, ax=ax, color=colors, width=0.75)


for container in ax.containers:
    vals = container.datavalues
    
    labels = [f"${v:,.0f}" if v > 10000 else "" for v in vals]
    
    ax.bar_label(
        container, 
        labels=labels, 
        label_type='center', 
        fontsize=6.5, 
        fontweight='bold', 
        color='white'
    )


for i, total in enumerate(totals):
    ax.text(
        total + 4000,     # Position text slightly to the right of the bar end
        i,                # Vertical position matching the bar index
        f"Total: ${total:,.0f}", 
        va='center', 
        ha='left', 
        fontsize=10.5, 
        fontweight='bold', 
        color='#2D3748'
    )

for spine in ['top', 'right', 'bottom', 'left']:
    ax.spines[spine].set_visible(False)

plt.title('Total Revenue (Sales) Breakdown by Industry and Segment', fontsize=16, fontweight='bold', pad=25, color='#1A202C')
plt.xlabel('Total Sales ($)', fontsize=11, labelpad=10, fontweight='bold', color='#4A5568')
plt.ylabel('Industry', fontsize=11, labelpad=10, fontweight='bold', color='#4A5568')

ax.get_xaxis().set_visible(False)

ax.set_xlim(right=totals.max() * 1.15)

plt.legend(title='Customer Segment', bbox_to_anchor=(1.02, 1), loc='upper left', frameon=False, fontsize=10)
plt.tight_layout()
plt.show()
```
