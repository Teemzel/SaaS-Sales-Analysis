```python
discount_diag = df.groupby('Discount')[['Sales', 'Profit']].mean().reset_index()
discount_diag['Profit_Margin_%'] = (discount_diag['Profit'] / discount_diag['Sales']) * 100
discount_diag['Sales'] = discount_diag['Sales'].map('{:.2f}'.format)
discount_diag['Profit'] = discount_diag['Profit'].map('{:.2f}'.format)
discount_diag['Profit_Margin_%'] = discount_diag['Profit_Margin_%'].map('{:.2f}'.format)
print("Average Profit Margin by Discount Level")
display(discount_diag.sort_values(by='Discount'))
```

```python
# Visualize the Discount-Profit Tipping Point
discount_diag['Profit_Margin_%'] = pd.to_numeric(discount_diag['Profit_Margin_%'], errors='coerce')
discount_diag = discount_diag.dropna(subset=['Profit_Margin_%'])
plt.figure(figsize=(10, 5))
sns.lineplot(data=discount_diag, x='Discount', y='Profit_Margin_%', marker='o', color='red', linewidth=2)
plt.axhline(0, color='black', linestyle='--', alpha=0.7)
for idx, row in discount_diag.iterrows():
    x = row['Discount']
    y = row['Profit_Margin_%']
    # Push labels slightly above positive points, and below negative points to prevent overlapping the line
    offset = 4 if y >= 0 else -12
    
    plt.annotate(
        f"{y:.1f}%", 
        (x, y),
        textcoords="offset points", 
        xytext=(0, offset), 
        ha='center', 
        fontsize=9, 
        fontweight='bold',
        color='darkred' if y < 0 else 'darkgreen'
    )
plt.title('The Tipping Point: How Discounts Destroy Profit Margin', fontsize=14)
plt.xlabel('Discount Percentage Applied')
plt.ylabel('Average Profit Margin (%)')
#plt.ylim(discount_diag['Profit_Margin_%'].min() - 20, discount_diag['Profit_Margin_%'].max() + 20)
plt.tight_layout()
plt.show()
```

Diagnostic Insight: The Discount Tipping Point
Insights
The 20% Break-Even Wall: Discounts up to 20% maintain positive margins (peaking at a healthy 29.5% with no discount). 
The Toxic Zone (≥ 30%): The moment discounts reach 30%, transactions cross the break-even threshold into negative territory (-10.1% margin) and rapidly spiral.
Catastrophic Value Destruction: At extreme promotional tiers like an 80% discount, the margin collapses to a staggering -180.0%, meaning the company loses nearly twice the sales value on every contract closed.

Strategic Recommendations
1. Enforce a Strict 20% Hard Cap: Mandate a firm ceiling in your sales CRM. Sales representatives must not exceed a 20% discount without multi-step executive approval.
2. Eliminate Loss-Leader Tactics: Stop the use of heavy discounting (30% to 80%) intended to volume-chase top-line revenue, as it severely burns bottom-line capital.
3. Transition to Value Add-ons: Train sales teams to protect margins by offering free sandboxed trials, onboarding support, or extra user seats rather than slashing contract pricing.

```python
# SAAS METRIC: B2B Account Health & Churn Risk Tracker

# Establish the "Current Date" as the maximum date in the dataset
snapshot_date = df['Order Date'].max()

saas_health = df.groupby(['Customer ID', 'Customer', 'Industry']).agg(
    Last_Active_Date=('Order Date', 'max'),
    Total_Lifetime_Spend=('Sales', 'sum')
).reset_index()

saas_health['Days_Since_Last_Active'] = (snapshot_date - saas_health['Last_Active_Date']).dt.days

def assign_health_tier(days):
    if days <= 30:
        return ' Active '
    elif days <= 60:
        return ' Needs Attention '
    else:
        return ' 60 Days Inactive '

saas_health['Account_Status'] = saas_health['Days_Since_Last_Active'].apply(assign_health_tier)

print("SaaS Customer Portfolio Health Summary")
display(saas_health['Account_Status'].value_counts().to_frame())

print("\nTop 5 Critical At-Risk Enterprise Accounts")

at_risk_df = saas_health[saas_health['Account_Status'] == ' 60 Days Inactive ']

# Sort and display the highest-value accounts that are slipping away
display(at_risk_df.sort_values(by='Total_Lifetime_Spend', ascending=False)
        .style.format({'Total_Lifetime_Spend': '${:,.2f}'}))
```
